```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
def get_states(dept):
    url = "https://gra108.aca.ntu.edu.tw/regchk/stu_query.asp"
    payload = {
        'DEP': dept,
        'qry': '查詢',
    }
    # Send a POST request with the specified parameters
    response = requests.post(url, data=payload)
    return response

def process_data(res, fname):
    s = BeautifulSoup(res.text, 'lxml')
    s = s.find("table")
    s = s.find_all("tr")
    ret = {
        "name": list(),
        "priority": list(),
        "choice": list(),
        "result": list()
    }
    count = 1
    for row in s[1: ]:
        data = row.find_all("td")
        data = [x.text for x in data][2:-2]
        ret["name"].append(data[0])
        pri = data[1]
        if "正取" in pri:
            ret["priority"].append(0)
        else:
            ret["priority"].append(pri[2: -2])
        choice = data[2]
        if "志願" in choice:
            ret["choice"].append(choice[2:])
        else:
            ret["choice"].append(-1)
        ret["result"].append("TBD")
    df = pd.DataFrame(ret)
    df.to_csv(f"raw_data/{fname}", index=False)
    return len(df)

```

```python
dfdct = dict()
for dpt in cleanedDpt:
    file_path = f"data/{dpt}.csv"
    df = pd.read_csv(file_path)
    dfdct[dpt] = df
```

```python
def saveDct(dct):
    for key, df in dct.items():
        df.to_csv(f"data/{key}.csv", index=False)
def getQueryDF(dct):
    ret = list()
    for key, df in dct.items():
        ndf = df.copy()
        ndf['department'] = key
        ret.append(ndf)
    return pd.concat(ret, ignore_index=True)
```

```python
for key, df in dfdct.items():
    df.loc[df["choice"] == -1, 'result'] = "Forfeit"
    df.loc[(df["choice"] == 1) & (df["priority"] == 0), 'result'] = "Enrolled"
saveDct(dfdct)
```

```python
QUOTA = {
    x: len(dfdct[x][dfdct[x]["priority"] == 0]) for x in cleanedDpt
}
```

```python
# 上之後把後面的志願序釋放
def release(name, c):
    # get the data
    query = getQueryDF(dfdct)
    choices = query[query["name"] == name].sort_values(by="choice")
    released = choices[choices["choice"] > c]
    processing =released[released["result"] == "TBD"]
    for dep in list(processing["department"]):
        # print(f"Releasing {name}'s choice on dept => {dep}")
        dfdct[dep].loc[dfdct[dep]["name"] == name, 'result'] = "Released"
    saveDct(dfdct)

        
```

```python
import numpy as np
def process(department):
    if QUOTA[department] == 0:
        print(f"{department} 根本沒收人...")
        return
    # get the department
    df = dfdct[department]
    last_enroll_idx = df.loc[df["result"] == "Enrolled"].index.max() + 1
    if(last_enroll_idx == len(df)):
        print(f"{department} 收完了...")    
        return
    if np.isnan(last_enroll_idx):
        print(f"{department} 從缺...")
        return 
    enrolled = df.iloc[:last_enroll_idx]
    counts = enrolled["result"].value_counts()
    counts["TBD"] = counts.get("TBD", 0)
    counts["Enrolled"] = counts.get("Enrolled", 0)
    print(f'Before: {department} => {counts["Enrolled"] + counts["TBD"]}/{QUOTA[department]}')
    left = QUOTA[department] - counts["Enrolled"] - counts["TBD"]
    left = min(len(df) - last_enroll_idx, left)
    flag = False
    if (left):
        
        lucky = df.iloc[last_enroll_idx:][
            (df.iloc[last_enroll_idx:]["result"] != "Forfeit") & (df.iloc[last_enroll_idx:]["result"] != "Released")
        ].iloc[:left]
        # print(lucky)
        e = lucky[lucky["choice"] == 1]["name"]
        for n in list(e):
            dfdct[department].loc[dfdct[department]["name"] == n, "result"] = "Enrolled"
        for _, person in lucky.iterrows():
            flag = True
            release(person["name"], person["choice"])
    return flag
# process("5060土木工程學系碩士班水利工程組")
```
